Problem, Metthod, and Options classes.

```python
# problem, method, and option classes
class Problem:
    def __init__(self, name, x0, compute_f, compute_g, compute_H=None):
        self.name = name
        self.x0 = x0.flatten()
        self.compute_f = lambda x: np.array(compute_f(x.reshape(-1, 1))).item()
        self.compute_g = lambda x: np.array(compute_g(x.reshape(-1, 1))).flatten()
        if compute_H:

            def safe_H(x):
                H_raw = compute_H(x.reshape(-1, 1))
                if isinstance(H_raw, np.ndarray) and H_raw.ndim == 0:
                    H_raw = H_raw.item()
                if hasattr(H_raw, "toarray"):
                    return H_raw.toarray()
                return np.array(H_raw, dtype=float)

            self.compute_H = safe_H
        else:
            self.compute_H = None


class Method:
    def __init__(self, name):
        self.name = name


class Options:
    def __init__(self):
        self.term_tol = 1e-6
        self.max_iterations = 1000
        self.c1 = 1e-4
        self.c2 = 0.9
        self.tau = 0.5
        self.backtracking_alpha = 1.0
        self.initial_alpha = 1.0
        self.alpha_low = 0.0
        self.alpha_high = 1000.0
        self.c = 0.5
        self.max_ls_iter = 100
        self.initial_radius = 1.0
        self.max_iterations_CG = 20
        self.term_tol_CG = 1e-5
```

Define a function to avoid repeating code for problem solving

```python
def run_algorithm(problem, method, options):
    print(f"Problem Solution Result: {method.name} on {problem.name}")
    t0 = time.time()
    try:
        x_opt, f_opt = optSolver_Fire_Horse(problem, method, options)
        t_elapsed = time.time() - t0

        print(f"Final f: {f_opt:.6f}")
        print(f"Time: {t_elapsed:.4f} sec")
        print(f"Final x: {np.round(x_opt, 4)}\n")
        return x_opt, f_opt

    except Exception as e:
        print(f"Error: {str(e).splitlines()[0]}\n")
        return None, None
```
